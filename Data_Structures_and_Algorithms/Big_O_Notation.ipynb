### Why not compute time complexity by counting milliseconds? 

Because in your computer it can take X miliseconds by in another it could take Y miliseconds. 

It is better if we could express time, or space requirements, as a function.

```python
import time
from matplotlib import pyplot as plt
import numpy as np

def square_array(array):
    for i, _ in enumerate(array):
        array[i] = array[i]**2
    return array

def how_long(array, func):
    t1 = time.time()
    func(array)
    t2 = time.time()
    return t2 - t1

n_list = np.array(range(100, 1100, 10))
time_executions = np.array([how_long([2]*n, square_array) for n in n_list])

plt.plot(n_list, time_executions) 
plt.xlabel('array size') 
plt.ylabel('execution time - ms') 
plt.title('Complexity Time') 
  
# function to show the plot 
plt.show() 
```

In the plot above we can see that the function *square_array* has a linear trend. Let's see:

```python
# m = slope, b=intercept
m, b = np.polyfit(n_list, time_executions, 1)

plt.plot(n_list, time_executions)
plt.plot(n_list, m*n_list + b)
plt.xlabel('array size') 
plt.ylabel('execution time - ms') 
plt.title('Complexity Time');
```

```python
print(m, b)
```

So we can state that the function *square_array* has a linear complexity. 

How express that in the Big O notation?

### Roles in Big O Notation

1. **Keep fastest growing term**: so $(m.n + b)$ becomes $m.n$
2. **Drop constants**: $m.n$ becomes  $n$

So as the *square_array* function has linear time complexity, we say it has O(n) complexity in big O notation. 

### If I have a function with two *for loops*?

```python
def function_with_2_for_loops(array):
    for i, _ in enumerate(array):
        for i, _ in enumerate(array):
            pass

time_executions = np.array([how_long([2]*n, function_with_2_for_loops) for n in n_list])

plt.plot(n_list, time_executions)
plt.xlabel('array size') 
plt.ylabel('execution time - ms') 
plt.title('Complexity Time');
```
